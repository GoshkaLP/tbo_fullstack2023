### Создание оснвовного датафрейма, из которого будут созданы производные

```python
df1 = df.rename(columns=new_columns)
```

```python
df1 = df1.dropna(subset=['name', 'yandex_object_coordinate_x', 'yandex_object_coordinate_y'])
```

```python
df1['obj_id'] = df1['obj_id'].astype(int)
```

```python
df1.to_csv('new_data.csv')
```

### Создание датафрейма df_loc с основной информацией об объектах и их местоположении

```python
df_loc = df1.loc[:, ['obj_id', 'name', 'federal_subject', 'address', 
                     'yandex_object_coordinate_x', 'yandex_object_coordinate_y']].rename(columns={'yandex_object_coordinate_x':'longitude',
                                                                                                 'yandex_object_coordinate_y':'latitude'})
```

```python
df_loc
```

### Создание датафрейма df_info со всей информацией об объектах

```python
df_info = df1.drop(['name', 'name_en',
    'yandex_object_coordinate_x', 'yandex_object_coordinate_y',
                   'yandex_map_scale', 'yandex_center_coordinate_x', 'yandex_center_coordinate_y',
                   'mini_coordinate_x', 'mini_coordinate_x', 'mini_coordinate_y', 'photo', 'object_photo_gallery_url',
                   'video', 'panoramas', 'web_translations', 'other_materials'], axis=1)
```
